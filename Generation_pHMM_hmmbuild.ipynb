Generate the profile hmm (saved as profile.hmm) from the uploaded FASTA

```python
!hmmbuild profile.hmm conservative-Lnb-alignedFASTA-eval-e^-10-DUF4105-domain-for-consensus-2-22-26.afa
```

-h can be used to check the hmmemit functions as well as modify the desired output

```python
!hmmemit -h
```

Now, using the profile.hmm, generate the majority-rule consensus sequence

```python
!hmmemit -c profile.hmm
```
